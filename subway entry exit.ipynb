# 월별 승하차 csv파일 병합(merge)

```python
import pandas as pd
import numpy as np
import glob
import os
import sys
import calendar
import datetime
import time
```

## 경로에 있는 csv파일을 새로운 csv파일로 합침

```python
input_path = sys.argv[1]
output_file = sys.argv[2]

all_files = glob.glob(os.path.join(input_path,'C:/Python/project01/subway/CARD_*.csv'))

all_data_frames = []

for file in all_files:
    data_frame = pd.read_csv(file, index_col=None, encoding = 'utf-8')
    all_data_frames.append(data_frame)

data_frame_concat = pd.concat(all_data_frames, axis=0, ignore_index=True)
data_frame_concat.to_csv("subway_merge", index = False) # 병합한 결과를 csv로 내보내기
```

## df에 주차요일, 년주차요일, 승하차총승객수 추가

```python
datelist=[]
datelist2=[]
for i in range(len(df)):
    date2=str(date[i])
    year=date2[:4]
    month=date2[4:6]
    day=date2[6:8]
    da=datetime.date(int(year),int(month),int(day)).isocalendar()[1:]
    da2=datetime.date(int(year),int(month),int(day)).isocalendar()
    datelist.append(da)
    datelist2.append(da2)
datelist
df["주차요일"] = datelist
df["년주차요일"] = datelist2
df["승하차총승객수"] = df["승차총승객수"] + df["하차총승객수"]
df.head()
```

## 승하차총승객수 내림차순

```python
sums=df["승하차총승객수"].groupby([df["년주차요일"],df["주차요일"],df["역명"]]).sum()
sums2=sums.reset_index(name="승하차총승객수").reindex(columns=["년주차요일","주차요일","역명","승하차총승객수"])
date6=pd.DataFrame(sums2)
means=date6["승하차총승객수"].groupby([date6["주차요일"],date6["역명"]]).mean()
means2=means.reset_index(name="승하차총승객수").reindex(columns=["주차요일","역명","승하차총승객수"])
obj=pd.DataFrame(means2)
obj2=obj.sort_values(by=['주차요일','승하차총승객수'], ascending=[True, False])
obj2
```
