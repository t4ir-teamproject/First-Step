```python
import pandas as pd
import glob
import os
import sys
```

```python
input_path = sys.argv[1]
output_file = sys.argv[2]

all_files = glob.glob(os.path.join(input_path,'C:/Users/Administrator/subway15-19/CARD_SUBWAY_MONTH_2016*.csv'))

all_data_frames = []

for file in all_files: 
    data_frame = pd.read_csv(file, index_col=None, encoding='utf-8') 
    all_data_frames.append(data_frame)

data_frame_concat = pd.concat(all_data_frames, axis=0, ignore_index=True) 
data_frame_concat.to_csv("2016_subway_merge.csv", index = False)
```
